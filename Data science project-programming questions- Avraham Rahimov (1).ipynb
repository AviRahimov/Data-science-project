```python
#Decimal to Binary
dec = int(input("write a decimal number: "))
str1 = ""
st = ""
while(dec!=0):
    st = st + str(dec%2)
    dec = dec//2
for i in range(len(st))[::-1]:
    str1 = str1 + st[i]
str1
```

```python
#Decimal to Octal
dec = int(input("write a decimal number: "))
str1 = ""
st = ""
while(dec!=0):
    st = st + str(dec%8)
    dec = dec//8
for i in range(len(st))[::-1]:
    str1 = str1 + st[i]
str1
```

```python
#Decimal to Hexadecimal
dec = int(input("write a decimal number: "))
str1 = ""
st = ""
while(dec!=0):
    if dec%16 == 10:
        st = st + 'A'
    elif dec%16 == 11:
        st = st + 'B'
    elif dec%16 == 12:
        st = st + 'C'
    elif dec%16 == 13:
        st = st + 'D'
    elif dec%16 == 14:
        st = st + 'E'
    elif dec%16 == 15:
        st = st + 'F'
    else:
        st = st + str(dec%16)
    dec = dec//16
for i in range(len(st))[::-1]:
    str1 = str1 + st[i]
str1
```
